### Informações importantes
- O arquivo possui $1.260.000$ pontos, como a órbita do planeta tem duração de $45.000$ segundos, logo, temos que em um segundo foram obtidos 28 pontos, em média.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arquivo = np.load("posicoes.npy")

coords_x = arquivo[:, 0]
coords_y = arquivo[:, 1]
```

```python
def ponto_no_tempo (coord, t):
    ''' Retorna o ponto do objeto no tempo t segundos
    
    Parâmetros:
    coord : coordenadas do arquivo
    t : tempo em segundos
    
    Retornos:
    ponto : ponto do objeto no tempo t'''

    n = (int) (t*28)

    try:
        return coord[n].copy()
    
    except:
        return [0, 0]
```

```python
x_values, y_values = [], []

origem = [0, 0]
point_zero = ponto_no_tempo(arquivo, 0)

for t in range(30):
    point = ponto_no_tempo(arquivo, t*50)

    x_values.append(point[0])
    y_values.append(point[1])

plt.plot([0], [0], 'ko', linewidth=10)

plt.plot(coords_x, coords_y, 'g',linewidth=3)
plt.plot(x_values, y_values, 'r')
plt.plot(x_values, y_values, 'bo')

plt.show()
```
